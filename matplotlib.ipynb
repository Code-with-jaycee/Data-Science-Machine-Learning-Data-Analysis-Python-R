Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(dpi=400)

x = np.linspace(20, 25, 10)
y = x**5
plt.plot(x, y,c="r", lw=0.4, marker="o", markerfacecolor="red")
plt.grid(True)
plt.title("Title")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

# Bargraph

```python
plt.figure(dpi=400)

x = np.arange(0, 6, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.grid(True)
plt.title("Title")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(dpi=600)

Month = ["jan", "Feb", "Mar", "Apr", "May"]
sales = [50, 60, 55, 75, 65]

plt.barh(Month, sales)

plt.title("Monthly sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
